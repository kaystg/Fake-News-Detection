**DATA PREPROCESSING**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Preprocessing and analysis of News column**

```python
from tqdm import tqdm
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from wordcloud import WordCloud
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def preprocess_text(text_data):
    preprocessed_text = []
    for sentence in tqdm(text_data):
        sentence = re.sub(r'[^\w\s]', '', sentence)
        preprocessed_text.append(' '.join(
            lemmatizer.lemmatize(token.lower())
            for token in str(sentence).split()
            if token not in stopwords.words('english')))
    return preprocessed_text
```

```python
preprocessed_review = preprocess_text(data['text'].values)
data['text'] = preprocessed_review
```

```python
# Real
consolidated = ' '.join(
    word for word in data['text'][data['class'] == 1].astype(str))
wordCloud = WordCloud(width=1600,
                      height=800,
                      random_state=21,
                      max_font_size=110,
                      collocations=False)
plt.figure(figsize=(15, 10))
plt.imshow(wordCloud.generate(consolidated), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer


def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx])
                  for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1],
                        reverse=True)
    return words_freq[:n]


common_words = get_top_n_words(data['text'], 20)
df1 = pd.DataFrame(common_words, columns=['Review', 'count'])

df1.groupby('Review').sum()['count'].sort_values(ascending=False).plot(
    kind='bar',
    figsize=(10, 6),
    xlabel="Top Words",
    ylabel="Count",
    title="Bar Chart of Top Words Frequency"
)
```

**Converting text into Vectors**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

x_train, x_test, y_train, y_test = train_test_split(data['text'],
                                                    data['class'],
                                                    test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer(ngram_range=(1, 3), max_features=5000)

x_train = vectorization.fit_transform(x_train)
x_test = vectorization.transform(x_test)
```

**Model training, Evaluation, and Prediction**

```python
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for Logistic Regression
param_grid_log_reg = {'C': [0.1, 1, 10], 'penalty': ['l2', 'l1'],'solver': ['liblinear']}
log_reg_grid = GridSearchCV(LogisticRegression(), param_grid_log_reg, cv=5)
log_reg_grid.fit(x_train, y_train)

# Get the best Logistic Regression model after tuning
model_lr = log_reg_grid.best_estimator_


# testing the model
print(accuracy_score(y_train, model_lr.predict(x_train)))
print(accuracy_score(y_test, model_lr.predict(x_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)

# testing the model
print(accuracy_score(y_train, model_dt.predict(x_train)))
print(accuracy_score(y_test, model_dt.predict(x_test)))
```

```python
from sklearn.ensemble import StackingClassifier

stacking_model = StackingClassifier(
    estimators=[('lr', LogisticRegression()), ('dt', DecisionTreeClassifier())],
    final_estimator=LogisticRegression()
)

# 4. Train the model on the training data
stacking_model.fit(x_train, y_train)

# 5. Evaluate the model
print(accuracy_score(y_train, stacking_model.predict(x_train)))
print(accuracy_score(y_test, stacking_model.predict(x_test)))
```

```python
# Confusion matrix of Results from Decision Tree classification
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, model_dt.predict(x_test))

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])

cm_display.plot()
plt.show()
```

```python
# Confusion matrix of Results from Logistic Regression classification
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, model_lr.predict(x_test))

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])

cm_display.plot()
plt.show()
```

```python
def classify_news(text_input):
    # Vectorize the input text
    input_vec = vectorization.transform([text_input])

    # Predict with both models
    log_reg_result = model_lr.predict(input_vec)[0]
    decision_tree_result = model_dt.predict(input_vec)[0]

    # Map predictions to labels
    log_reg_label = "Fake" if log_reg_result == 0 else "True"
    decision_tree_label = "Fake" if decision_tree_result == 0 else "True"

    return log_reg_label, decision_tree_label

# Sample input for testing
sample_input = "Barcelona balances security and freedom after deadly attacks."
log_reg_class, dt_class = classify_news(sample_input)

log_reg_class, dt_class
```
